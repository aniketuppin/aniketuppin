```python
import numpy as np 
import scipy.io.wavfile as wav
import matplotlib.pyplot as plt
import scipy.signal as sig


n=np.arange(-10,9)
t = np.arange(0,5,(1.0/2))
x_re= -np.sin(3*np.pi*n/16)
x_img=np.cos(3*np.pi*n/16)
plt.stem(n,x_re)
plt.title("real part")
plt.show()
plt.stem(n,x_img)
plt.title("imaginary part")
plt.show()

b=20*[1]
a=10*[0]+[1]+ 9*[0]
c=np.arange(-np.pi,np.pi,2*np.pi/4096)
w1,h1=sig.freqz(b,a,c)
h1db=20*np.log10(abs(h1))
plt.plot(w1/(2*np.pi),h1db)
plt.title("magnitude spectrum")
plt.show()

angles = np.angle(h1,deg=True)
plt.plot(w1/(2*np.pi), angles, 'g')
plt.title("phase spectrum")
plt.show()
```

the magnitude spectrum is even function and phase spectrum is odd function
