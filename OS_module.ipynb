```python
def most_common_word():
    txtfile=glob.glob("*.txt")
    count=dict()
    
    for txt in txtfile:
        with open(txt) as f:
            line=f.readline()
            while (line!=""):
                words=line.split()
                for word in words:
                    word=word.replace(',','').replace('.','').lower()
                    if word not in count:
                        count[word]=1
                    else:
                        count[word]+=1
                line=f.readline()
    maxkey=max(count,key=count.get)
    maxvalue=max(count.values())
    
    print("the most appeared word is ",maxkey)
    print("frequency of ",maxkey," is ",maxvalue)
        
```
