Nearest Neighbour Algorithm

```python
class nearest _neighbour:
    def __init__(self)
    pass

def  train(self,X,y):
    self.xtrain=X
    self.ytrain=y
    
def test(self,X):
    num=X[0].shape
    yprd=np.zeros(num,dtype=self.ytrain.dtype)

    for i in range(num):
    dist = np.sum(np.abs(self.xtrain - X[i,:]),axis=1)
    #print(dist.shape)
    min_index=np.argmin(dist)
    yprd[i] =self.ytrain[min_index]
    return yprd
```
