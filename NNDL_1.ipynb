```python
# Write python code for plotting the following activation functions for the input x in the range of -20 to 20.

#(a) Sigmoid (b) TanH (c) Softsign (d) ReLU (e) Softmax

#Write python code for the derivative of the following activation functions and their plots for the input x in the range of -20 to 20.

#(a) Sigmoid (b) TanH (c) Softsign (d) ReLU

import numpy as np
import matplotlib.pyplot as plt

x=np.arange(-20,20,0.05)

#sigmoid function and its derivative
sigmoid_x= 1/(1 + np.exp(-x))
del_sigmoid_x= sigmoid_x*(1-sigmoid_x)
#print(sigmoid_x)
plt.plot(x,sigmoid_x,label="sigmoid_x")
plt.plot(x,del_sigmoid_x,label="d/dx(sigmoid_x)")
plt.legend()
plt.grid()
plt.xlabel("x")
plt.ylabel("activation_fn")
plt.title("SIGMOID")

#tanh and its derivative
tanh_x=(np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
del_tanh_x= 1-(tanh_x**2)
plt.figure(2)
plt.plot(x,tanh_x,label="tanh_x")
plt.plot(x,del_tanh_x,label="d/dx(tanh_x)")
plt.legend()
plt.grid()
plt.xlabel("x")
plt.ylabel("activation_fn")
plt.title("TANH")

#softsign and its derivative
softsign_x= x/(1 + np.abs(x))
del_softsign_x= 1/(1 + np.abs(x))**2
plt.figure(3)
plt.plot(x,softsign_x,label="softsign_x")
plt.plot(x,del_softsign_x,label="d/dx(softsign_x)")
plt.legend()
plt.grid()
plt.xlabel("x")
plt.ylabel("activation_fn")
plt.title("SOFTSIGN")

#relu and its derivative
relu_x= np.maximum(0,x)
del_relu_x=np.where(x>0,1,0)
#derivative not defined at 0
plt.figure(4)
plt.plot(x,relu_x,label="relu_x")
plt.plot(x,del_relu_x,label="d/dx(relu_x)")
plt.legend()
plt.grid()
plt.xlabel("x")
plt.ylabel("activation_fn")
plt.title("ReLU")

#softmax
softmax_x=np.exp(x)/(np.sum(np.exp(x)))
plt.figure(5)
plt.plot(x,softmax_x,label="softmax_x")
plt.legend()
plt.grid()
plt.xlabel("x")
plt.ylabel("activation_fn")
plt.title("SOFTMAX")
```

```python
#import matplotlib
#Assume we have a 2-input neuron and has the following parameters: w = [1, 1] ; b=2 .
#Write a python code for calculating the feed-forward output of neural networks by using following activation functions for input x=[4, 5].

X=np.array([4,5])
w=[1,1]
b=2
print(type(X))
# z=wx + b
def sigmoid(x):
    return 1/(1 + np.exp(-x))
def relu(x):
    return np.maximum(0,x)
def tanh(x):
    return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
def softsign(x):
    return x/(1 + np.abs(x))


def f_z(X,w,b,activation_fn):
    Z=np.dot(w,X) + b
    Z=np.expand_dims(Z,axis=-1)
    #print(Z)
    #print(type(Z))
    f=activation_fn(Z)
    return f


sigmoid_out=f_z(X,w,b,sigmoid)
print("output from sigmoid activation function: ",sigmoid_out)
relu_out=f_z(X,w,b,relu)
print("output from relu activation function: ",relu_out)
tanh_out=f_z(X,w,b,tanh)
print("output from tanh activation function: ",tanh_out)
softsign_out=f_z(X,w,b,softsign)
print("output from softsign activation function: ",softsign_out)

x_axis =np.array(["sigmoid","relu","tanh","softsign"])
y_axis=np.concatenate([sigmoid_out,relu_out,tanh_out,softsign_out])
c=['red','blue','green','black']
p=plt.bar(x_axis,y_axis,color=c)

plt.grid
```
